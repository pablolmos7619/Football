### Agrupamos las variables por EQUIPOS

```python
agrupa=medios.groupby(['team']).sum()
```

```python
agrupa.reset_index(level=0, inplace=True)
```

## Factor1. Contundencia en el duelo

```python
agrupa['duelos_ganados'] = agrupa['duels_won']+agrupa['duels_ground_won']+agrupa['duels_aerial_won']
agrupa['duelos_ganados_total'] = agrupa['duels_total']+agrupa['duels_ground_total']+agrupa['duels_aerial_total']
agrupa['contundencia']=agrupa['duelos_ganados'] /agrupa['duelos_ganados_total']
```

## Factor2. Capacidad de asociación

```python
agrupa ['paseok']= agrupa['passes_short_successful'] +agrupa['passes_long_successful'] + agrupa['passes_own_half_successful']+ agrupa['passes_opposition_half_successful']
agrupa ['pasetotal']= agrupa['passes_short_successful'] +agrupa['passes_short_unsuccessful'] +agrupa['passes_long_successful']+agrupa['passes_long_unsuccessful']  + agrupa['passes_own_half_successful'] + agrupa['passes_own_half_unsuccessful']+ agrupa['passes_opposition_half_successful'] + agrupa['passes_opposition_half_unsuccessful']
agrupa ['asociacion']=agrupa ['paseok']/agrupa ['pasetotal']
```

## Factor3. Acierto de Cara a Puerta

```python
agrupa ['Acierto_Cara_Puerta']=agrupa['goals']/agrupa['shots_total']
```

## Factor4. Capacidad de Dribbling y Asistencia

```python
agrupa ['dribSuccess']=agrupa ['dribbles_successful'] / (agrupa ['dribbles_unsuccessful']+ agrupa['dribbles_successful'])
```

```python
agrupa ['Asistencia']=(agrupa ['keypasses']+agrupa ['assists']+agrupa ['assists_intentional']+agrupa ['assists_second_goal'])/agrupa['passes_total']
```

```python
agrupa['dribItention']=agrupa ['dribSuccess']+(agrupa ['Asistencia'])
```

## Factor5- Portero Decisivo

```python
agrupa['intervenciones']=agrupa ['clearances']+ agrupa['interceptions']+ agrupa['blocks']
agrupa['indice_paradas']=1-(agrupa['goals_conceded']/agrupa['intervenciones'])
```

```python
agrupa['pasesokgk']=agrupa['passes_open_play_successful']/agrupa['passes_open_play']
```

```python
agrupa['Geat_GK']=0.8*agrupa['indice_paradas']+0.2*agrupa['pasesokgk']
```

```python
agrupa
```

```python
factores=agrupa[['team','contundencia','asociacion','Acierto_Cara_Puerta','dribItention','Geat_GK']]
```

```python
factores['Score_Best_Team']=1/5*factores['contundencia']+1/5*factores['asociacion']+1/5*factores['Acierto_Cara_Puerta']+1/5*factores['dribItention']+1/5*factores['Geat_GK']
```

```python
class ComplexRadar():
    def __init__(self, fig, variables, ranges, n_ordinate_levels=5):
        angles = np.arange(0, 360, 360./len(variables))

        axes = [fig.add_axes([0.1,0.1,0.9,0.9], polar=True,
            label = "axes{}".format(i)) 
            for i in range(len(variables))]

        l, text = axes[0].set_thetagrids(angles, labels=variables)

        [[txt.set_fontweight('bold'),
              txt.set_fontsize(12),
              txt.set_position((0,-0.2))] for txt in text]

        for ax in axes[1:]:
            ax.patch.set_visible(False)
            ax.grid("off")
            ax.xaxis.set_visible(False)

        for i, ax in enumerate(axes):
            grid = np.linspace(*ranges[i], num=n_ordinate_levels)
            gridlabel = ["{}".format(round(x,2)) for x in grid]

            gridlabel[0] = "" # clean up origin
            ax.set_rgrids(grid, labels=gridlabel,angle=angles[i])

            ax.set_ylim(*ranges[i])
        
        # variables for plotting
        self.angle = np.deg2rad(np.r_[angles, angles[0]])
        self.ranges = ranges
        self.ax = axes[0]

    def plot(self, data, *args, **kw):
        sdata = self.scale_data(data, self.ranges)
        self.ax.plot(self.angle, np.r_[sdata, sdata[0]], *args, **kw)

    def fill(self, data, *args, **kw):
        sdata = self.scale_data(data, self.ranges)
        self.ax.fill(self.angle, np.r_[sdata, sdata[0]], *args, **kw)

    def scale_data(self, data, ranges):
        """scales data[1:] to ranges[0]"""
        for d, (y1, y2) in zip(data[1:], ranges[1:]):
            assert (y1 <= d <= y2) or (y2 <= d <= y1)
        x1, x2 = ranges[0]
        d = data[0]
        sdata = [d]
        for d, (y1, y2) in zip(data[1:], ranges[1:]):
            if y1 > y2:
                d = _invert(d, (y1, y2))
                y1, y2 = y2, y1
            sdata.append((d-y1) / (y2-y1) * (x2 - x1) + x1)
        return sdata
```
